# generateurs d'index

```python
import json

def generer_index_par_exercice(index):
    S = ""
    i = 1
    for key, val in index.items():
        S = S + '**' + '{:02}'.format(i) + ') [' + key.replace('_', ' ') + '](' + key + '.ipynb)' + '**  \n'
        S = S + '- Référence : ' + val[1][1:] + '  \n'
        S = S + '- Notions : ' + ', '.join(val[2:]).replace('_', ' ') + '  \n'
        S = S + '  \n'
        i = i + 1
    return S

with open('index.json', 'r') as json_file:
    index = json.load(json_file)
S = generer_index_par_exercice(index)
print(S)
```

```python
def recuperer_liste_notions(index):
    s = set()
    for key, val in index.items():
        s = s.union(val[2:])
    return list(s)

def generer_index_par_notion(index):
    notions = recuperer_liste_notions(index)
    S = ""
    for notion in notions:
        S = S + '**' + notion.replace('_', ' ') + '**  \n'
        for key, val in index.items():
            if notion in val:
                S = S + '[' + key.replace('_', ' ') + '](' + key + '.ipynb)' + ', '
        S = S[:-2] + '  \n'
        S = S + '  \n'
    return S

S = generer_index_par_notion(index)
print(S)
```
